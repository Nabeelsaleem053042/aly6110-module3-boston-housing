```python
%%R
# 1
df <- read.csv("BostonHousing.csv")

# 2
head(df, 5)

# 3
nrow(df)

# 4
str(df)

# 5
df2 <- subset(df, select = -b)

# 6
df_round <- as.data.frame(lapply(df2, function(x) {
  if (is.numeric(x)) {
    round(x, 2)
  } else {
    x
  }
}))

# 7
df_round$Age10 <- round(df_round$age * 1.10, 2)

# 8
hist(df_round$Age10,
     main = "Histogram of Age10",
     xlab = "Age10",
     ylab = "Frequency")

# 9
summary(df_round)

# 11
tail(df_round, 5)
```
