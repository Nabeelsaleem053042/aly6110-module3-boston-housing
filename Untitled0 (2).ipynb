```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, round as spark_round
import matplotlib.pyplot as plt

# Initialize Spark
spark = SparkSession.builder.appName("BostonHousing").getOrCreate()

# 1. Read CSV
df = spark.read.csv("BostonHousing.csv", header=True, inferSchema=True)

# 2. Show first 5
df.show(5)

# 3. Count observations
df.count()

# 4. Show schema
df.printSchema()

# 5. Drop column 'b'
df = df.drop('b')

# 6. Round all numerical columns
numeric_cols = [f.name for f in df.schema.fields if f.dataType.simpleString() in ['double', 'float']]
for col_name in numeric_cols:
    df = df.withColumn(col_name, spark_round(col(col_name), 2))

# 7. Create Age10 column
df = df.withColumn('Age10', spark_round(col('age') * 1.10, 2))

# 8. Plot histogram (need to convert to Pandas first)
age10_pandas = df.select('Age10').toPandas()
plt.hist(age10_pandas['Age10'], bins=20)
plt.title('Histogram of Age10')
plt.xlabel('Age10')
plt.ylabel('Frequency')
plt.show()

# 9. Summary statistics
df.describe().show()

# 10. Convert to Pandas
pandas_df = df.toPandas()

# 11. Show last 5 rows
pandas_df.tail(5)
```
